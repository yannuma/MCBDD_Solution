```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Parameters for following experiment ##
sensitivity = 0.99
spec1 = 0.99
spec2 = 0.999
spec3 = 0.9999
spec4 = 0.99999
prev = np.arange(0.00001, 0.5, 0.00001)

#Calculates the probability that Fred(or anyone else) is efected,when the test is positive
#Parameters:
#prev is the prevelance of the disease
#spec the specifity of the test
#sens the sensitivity of the test
def fred_infected(prev, spec, sens):
    #Bayes formula
    return (sens * prev )/(sens * prev + (1 - spec) * (1 - prev))
 
#calculating the probabilities
pfpos_spec1 = fred_infected(prev, spec1, sensitivity)
pfpos_spec2 = fred_infected(prev, spec2, sensitivity)
pfpos_spec3 = fred_infected(prev, spec3, sensitivity)
pfpos_spec4 = fred_infected(prev, spec4, sensitivity)

#Plot the results
plt.plot(prev, pfpos_spec1, label='Specifity = 99%')
plt.plot(prev, pfpos_spec2, label='Specifity = 99.9%')
plt.plot(prev, pfpos_spec3, label='Specifity = 99.99%')
plt.plot(prev, pfpos_spec4, label='Specifity = 99.999%')
plt.xlabel('Prevelance')
plt.ylabel('P(F+|t+)')
plt.title('Influence of Prevalence and Specifity, Sensitivity = 99%')
plt.legend()
plt.grid(True)
plt.show()
```

```python
## Integer Example to check the Results ##

#There are 100 People
#Prevalence is 10% -> 10 People have the disease (are positive), 90 People are negative
#Sensitivity is 90% -> This means from the 10 positive People, 9 are True Positive and 1 is False Negative
#Specifity is 90% -> This means from the 90 People without the disease, 81 are True Negative and 9 are False Positive

#The calculate now the probability use his formula True Positive / (True Positive + False Positive)
print("Prob Integer example:", 9 / (9 + 9))
print("Prob. from Function:",fred_infected(0.10, 0.9, 0.9))
print("Therefore my function should be correct")

#Check the result from 1a -> it's also the same
print("Result 1a: ",fred_infected(0.05, 0.995, 0.995))
```
